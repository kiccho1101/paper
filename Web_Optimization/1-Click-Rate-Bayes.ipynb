```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from typing import Callable
```

## Prior Distribution
Uniform distribution: $p(\theta) = 1$ 

```python
thetas = np.linspace(0, 1, 1001)
p = np.array([1 / len(thetas) for _ in thetas])
plt.plot(thetas, p)
plt.xlabel(r"$\theta$")
plt.ylabel(r"$p(\theta)$")
plt.show()
```

## Likelihood Function
- Bernoulli distribution: $p( r | \theta ) = \theta^r(1 - \theta)^(1-r)$

```python
likelihood = lambda r, thetas: thetas if r == 1 else (1 - thetas)
```

- $\theta = 0.5$ \
    $p(r|\theta=0) = 0.5^r * 0.5^(1-r)$
- $\theta = 0.3$ \
    $p(r|\theta=0) = 0.3^r * 0.7^(1-r)$

```python
theta = 1.0
for theta in np.linspace(0, 1, 10):
    ber = lambda r: theta if r == 1 else (1 - theta)
    plt.bar([0, 1], [ber(0), ber(1)])
    plt.xlabel(r"$r$")
    plt.ylabel(r"$p(r|\theta={})$".format(theta))
    plt.show()
```

## Posterior Distribution
$p(\theta|r) = p(r|\theta) * p(\theta)$

```python
def posterior(r: int, thetas: np.ndarray, likelihood: Callable, prior: np.ndarray) -> np.ndarray:
    lp = likelihood(r, thetas) * prior
    return lp / lp.sum()
```

```python
p = np.array([1 / len(thetas) for _ in thetas])
n_clicks = 3
n_nonclicks = 4
trials = [0] * n_nonclicks + [1] * n_clicks
for r in trials:
    p = posterior(r, thetas, likelihood, p)
    plt.plot(thetas, p)
    plt.xlabel(r"$\theta$")
    plt.ylabel(r"$p(\theta|r)$".format(theta))
    plt.show()
```
